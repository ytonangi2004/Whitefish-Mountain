```python
### (Whitefish mountain Ratio_of TerrainParks_To_Other ScatterPlot
### (WFM_TerrainParks_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")

print(df)
x = df["TerrainParks"]
y = df["ratio_of_TerrainParks_other_to_WFM"]
plt.xlabel('TerrainParks') 
plt.ylabel('Ratio of TerrainParks Other to WFM ') 
plt.title('TerrainParks Value of Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show() 
```
