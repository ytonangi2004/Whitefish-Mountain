```python
### (Whitefish mountain Ratio_of triple_To_Other ScatterPlot
### (WFM_AdultWeekday_To_Other Ratio ScatterPlot
### Please note that the lower/upper case notation, is followed as described in the metadata names 
### -------------------------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")
x = df["AdultWeekday"]
y = df["ratio_of_AdultWeekday_other_to_WFM"]
plt.xlabel('Adult Weekday Ticket in $') 
plt.ylabel('Ratio of Adult Weekday Ticket, Other to WFM ') 
plt.title('Adult Weekday Ticket in $ for Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show()
```
