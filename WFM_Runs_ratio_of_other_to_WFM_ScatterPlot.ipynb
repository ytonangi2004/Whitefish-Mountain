```python
### (Whitefish mountain Ratio_of Runs_To_Other ScatterPlot
### (WFM_Runs_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")

print(df)
x = df["Runs"]
y = df["ratio_of_Runs_other_to_WFM"]
plt.xlabel('Runs') 
plt.ylabel('Ratio of Runs Other to WFM ') 
plt.title('Runs Value of Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show() 
```
