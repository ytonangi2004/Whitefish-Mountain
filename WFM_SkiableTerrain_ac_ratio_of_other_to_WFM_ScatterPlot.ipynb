```python
### (Whitefish mountain Ratio_of triple_To_Other ScatterPlot
### (WFM_SkiableTerrain_ac_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")
x = df["SkiableTerrain_ac"]
y = df["ratio_of_SkiableTerrain_ac_other_to_WFM"]
plt.xlabel('Skiable Terrain in Acres') 
plt.ylabel('Ratio of Skiable Terrain in Acres Other to WFM ') 
plt.title('Skiable Terrain in Acers Value of Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show()
```
