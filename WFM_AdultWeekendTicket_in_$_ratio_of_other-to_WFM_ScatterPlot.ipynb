```python
### (Whitefish mountain Ratio_of triple_To_Other ScatterPlot
### (WFM_AdultWeekend_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")
x = df["AdultWeekend"]
y = df["ratio_of_AdultWeekend_other_to_WFM"]
plt.xlabel('Adult Weekend Ticket in $') 
plt.ylabel('Ratio of Adult Weekend Ticket, Other to WFM ') 
plt.title('Adult Weekend Ticket in $ for Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show()
```
