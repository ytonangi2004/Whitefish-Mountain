```python
### (Whitefish mountain Ratio_of triple_To_Other ScatterPlot
### (WFM_averageSnowfall_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")
x = df["averageSnowfall"]
y = df["ratio_of_averageSnowfall_other_to_WFM"]
plt.xlabel('Average Snowfall') 
plt.ylabel('Ratio of Average Snowfall Other to WFM ') 
plt.title('Average Snowfall Value of Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show()
```
