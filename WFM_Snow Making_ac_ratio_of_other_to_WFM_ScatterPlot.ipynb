```python
### (Whitefish mountain Ratio_of triple_To_Other ScatterPlot
### (WFM_Snow Making_ac_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")
x = df["Snow Making_ac"]
y = df["ratio_of_Snow Making_ac_other_to_WFM"]
plt.xlabel('Snow Making in Acres') 
plt.ylabel('Ratio of Snow Making in Acres Other to WFM ') 
plt.title('Snow Making in Acers Value of Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show()
```
