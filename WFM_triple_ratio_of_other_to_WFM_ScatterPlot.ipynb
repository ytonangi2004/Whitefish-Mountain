```python
### (Whitefish mountain Ratio_of triple_To_Other ScatterPlot
### (WFM_triple_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")
x = df["triple"]
y = df["ratio_of_triple_other_to_WFM"]
plt.xlabel('Triple') 
plt.ylabel('Ratio of Triple Other to WFM ') 
plt.title('Triple Value of Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show() 
```
