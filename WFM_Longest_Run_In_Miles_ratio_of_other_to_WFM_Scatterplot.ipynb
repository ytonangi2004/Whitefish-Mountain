```python
### (Whitefish mountain Ratio_of triple_To_Other ScatterPlot
### (WFM_LongestRun_mi_To_Other Ratio ScatterPlot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ski_data_with_ratios.csv")
x = df["LongestRun_mi"]
y = df["ratio_of_LongestRun_mi_other_to_WFM"]
plt.xlabel('Longest Run in Miles') 
plt.ylabel('Ratio of Longest Run in Miles Other to WFM ') 
plt.title('Longest Run in Miles Value of Whitefish Mountain Resort')

plt.scatter(x, y)
plt.show()
```
