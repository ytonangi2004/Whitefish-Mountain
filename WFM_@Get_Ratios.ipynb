```python
### WFM_@Get_Ratios.ipynb
### Whitefish Mountain Resort has several facilities better than its
### competitors. This script gets the ratios of the following eleven facilities to that of Whitefish Mountain(WFM)
### ( ski_data_A.csv ) + (WFM_@Get_Ratios.ipynb) ---> ("ski_data_with_ratios.csv")
### Please note that the lower/upper case notation, is followed as described in the metadata names 
### -------------------------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("ski_data_A.csv")
df.drop(df.columns[0], axis=1)
print(df.shape)

### print(df.base_elev == 4464 ) 
### print( df[ ( df.Name == “Whitefish Mountain Resort”) & (df.base_elev == 4464 ) ])
### print(df[['Name', 'base_elev', 'total_chairs']] )
### print(df.isnull().sum())
### new column defined as ratio of total_chairs in Whitefish to other resort

df["ratio_of_triple_other_to_WFM"] = round((df["triple"]/6),4)
df["ratio_of_surface_other_to_WFM"] = round((df["surface"]/3),4)
df["ratio_of_total_chairs_other_to_WFM"] = round((df["total_chairs"]/14),4)
df["ratio_of_Runs_other_to_WFM"] = round((df["Runs"]/105),4)
df["ratio_of_TerrainParks_other_to_WFM"] = round((df["TerrainParks"]/4),4)
df["ratio_of_LongestRun_mi_other_to_WFM"] = round((df["LongestRun_mi"]/3.3),4)
df["ratio_of_SkiableTerrain_ac_other_to_WFM"] = round((df["SkiableTerrain_ac"]/3000),4)
df["ratio_of_Snow Making_ac_other_to_WFM"] = round((df["Snow Making_ac"]/600),4)
df["ratio_of_averageSnowfall_other_to_WFM"] = round((df["averageSnowfall"]/333),4)
df["ratio_of_AdultWeekday_other_to_WFM"] = round((df["AdultWeekday"]/81),4)
df["ratio_of_AdultWeekend_other_to_WFM"] = round((df["AdultWeekend"]/81),4)

print(df.shape)
print(df)
print(df.columns)
### df.WFM_triple_To_Other = 6/df.triple
### print(df.loc[df.triple !=0])
### df.WFM_triple_To_Other'] = (6/(df.loc[df.triple!=0])

df.to_csv("ski_data_with_ratios.csv")
```
